Below we set up the dataframe with the Goodreads dataset for you. If you run this cell, you may get an error that some lines were skipped. Do not worry about that! For this studio, we want to focus on creating an effective

```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
```

With your dataframe set up, time to start creating visualizations!

1 Number of books published per year
2 Number of books published by each publisher (some publisher’s names are in non Latin scripts, such as Japanese and Russian so   Matplotlib may not know how to display these names, that is okay!)
3 Text ratings count versus reviews count

```python
# Use your Python skills to create a visualization here.

style.use('ggplot')
plt.rcParams['figure.figsize']=(30,10)
plt.xlabel('years',color='blue')
plt.ylabel('Number of books',color='blue')
plt.xticks(color='black')
plt.yticks(color='black')
plt.title('Number of books published each year',color='blue')
df['year'].value_counts()[:20].plot(kind='bar',color='blue')
```

```python
plt.figure(figsize=(20,8))
g = sb.countplot(x="year", data=df)
g.set_title("title")
```

```python
# Use your Python skills to create a visualization here.
style.use('ggplot')
plt.rcParams['figure.figsize']=(30,10)
plt.xlabel('publisher',color='black')
plt.ylabel('Number of books',color='black')
plt.xticks(color='black')
plt.yticks(color='black')
plt.title('Number of books published each year',color='black')
df['publisher'].value_counts()[:20].plot(kind='bar',color='orange')
```

```python
plt.rcParams['figure.figsize'] = (30,10) 

plt.xlabel('Years',color = 'blue')
plt.ylabel('Number of books',color = 'blue')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.title('Number of books published each year',color = 'blue')
df['publisher'].value_counts().head(10).sort_values(ascending=True).plot(kind='barh' , color = 'coral')
```

```python
sb.scatterplot('text_reviews_count','ratings_count', data = df, s = 200, linewidth = 3, edgecolor = 'black',color="lightgreen")
plt.title('Text ratings count versus reviews count', color = 'black', fontsize = 18)
plt.savefig('scatter_review.png')

plt.show()
```

```python
sb.regplot(x = 'text_reviews_count', y = 'ratings_count', data = df, color = 'green', scatter_kws = {'s':150,'alpha':0.4})
plt.title('Text ratings count versus reviews count', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('text_reviews_count', color = 'black')
plt.ylabel('ratings_count', color = 'black')
plt.savefig('reg_review_ratings.png')

plt.show()
```
